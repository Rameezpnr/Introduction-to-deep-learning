```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
from keras.layers.convolutional import Conv2D # to add convolutional layers
from keras.layers.convolutional import MaxPooling2D # to add pooling layers
from keras.layers import Flatten # to flatten data for fully connected layers
```

```python
#Read dataset into pandas dataframe
#importing pandas and numpy
import pandas as pd
import numpy as np
concrete_data=pd.read_csv('concrete_data.csv')
concrete_data.head(5)
```

```python
dataset = concrete_data.values
```

```python
#normalising the data
dataset_norm = []
for i in range(9):
    _mean = np.mean(dataset[:,i])
    _std = np.std(dataset[:,i])
    _norm = (dataset[:,i] - _mean)/_std
    dataset_norm.append(_norm)
    
print(dataset_norm)
```

```python
#Making transpose
dataset_normalised = np.asarray(dataset_norm).T
print(dataset_normalised)
```

```python
# split the concrete_data set into predictors (inputs) and target (output)
X = dataset_normalised[:,0:7]
y = dataset_normalised[:,8]
```

```python
#One hidden layer of 10 nodes, and a ReLU activation function
#Use the adam optimizer and the mean squared error as the loss function.
# Create the neural network in a function so we can use it multiple times in the
# subsequent sections
def baseline_model():
    #create the model
    model = Sequential()
    model.add(Dense(7, input_dim=7, activation='relu'))
    model.add(Dense(10,activation = 'relu'))
    model.add(Dense(1))
    #compile the model
    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    return model
model = baseline_model()
```

```python
# import scikit-learn in order to randomly split the data into a training and test sets
from sklearn.model_selection import train_test_split
#Splitting the data into a training and test sets by holding 30% of the data for testing.
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
X_train.shape
```

```python
# Added extra lines

# train the model for 50 epochs.
epochs = 50
model.fit(X_train, y_train, epochs=epochs, verbose=1)
```

```python
#evaluate the model on the test data.
loss_val = model.evaluate(X_test, y_test)
y_pred = model.predict(X_test)
loss_val
```

```python
# import the mean_squared_error function from Scikit-learn.
from sklearn.metrics import mean_squared_error
mean_square_error = mean_squared_error(y_test, y_pred)
mean = np.mean(mean_square_error)
standard_deviation = np.std(mean_square_error)
print(mean, standard_deviation)

#added lines ended
```

```python
mse = []
for i in range(50):
    model = baseline_model()
    model.fit(X_train,y_train, epochs = 50)
    y_pred = model.predict(X_test)
    mse.append(mean_squared_error(y_test,y_pred))
```

```python
mean = np.mean(mse)
mean
```

```python
std = np.std(mse)
std
```

```python
print("Mean squared errors for 50 regression models: mean = %.3f, std dev = %.3f" %(mean, std))
```
